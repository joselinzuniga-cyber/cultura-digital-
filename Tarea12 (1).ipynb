## ¿Qué es un valor booleano?
Los **booleanos** son un tipo de dato que representan valores de verdad, es decir si algo es verdadero o falso. Cundo usamos en el `if` la condicion tomaba un valor de verdad y dependiendo de su valor realizabamos alguna acción; es decir, los usamos para tomar decisiones en los programas.. En python tendremos los valores `True` para el valor de verdadero y `False` para el valor de falso.


```python
a = True
b = False
print("a:", a)
print("b:", b)
```

Nota que pasa si mandamos a pantalla condiciones parecidas a las que usamos en los ejercicios de las condicionales.

```python
print("Ejemplo uno: ",5<6)
print("Ejemplo dos: ",5>6)
print("Ejemplo tres: ",5==6)
```

## **Ciclo `while`**



El ciclo `while` repite un bloque **mientras** una condición sea verdadera.

```python
#Ejemplo:

#Definimos una variable llamada contador con el valor de 1.
contador = 1

#Mientras la variable tenga un valor menor a 5 vamos a repetir el código siguiente.
while contador < 5:
    #Mandamos a pantalla Contador seguido del valor que tiene la varibale contador.
    print("Contador:", contador)
    #Aumentamos el valor de contador
    contador = contador + 1
```

Nota que en el código pasado dentro del while la condición que se está evaluando es si el valor de la variable `contador` es menor a 5; y justo ese es el valor booleano, mientras sea cierto el valor de `contador < 5` es `True`; en el momento en que no se cumpla el valor de `contador < 5` será `False`.



Veamos otro ejmeplo, pero ahora le pasaremos explicitamente un valor boolenao a la condición del while.


```python
#Definimos una variable llamada activo con el valor boolenao True
activo = True
#Definimos una variable llamada número con el valor 0
numero = 0

#Mientras la variable activo tenga el valor de True realiza el siguiente código.
while activo:
    #Manda a pantalla Numero seguido del valor de la variable número.
    print("Número:", numero)
    #Aumenta el valor de la variable número en uno.
    numero = numero + 1
    #Si el valor de la variable numero es igual a 4 realiza lo siguiente.
    if numero == 4:
        #Cambia el valor de la varibale activo a False.
        activo = False
```
